# **Questão 1**

No lab anterior, foi realizado o tratamento de maiúsculas/minúsculas e palavras acentuadas. Neste lab, utilizarei a biblioteca nltk para realizar a tokenização filtrando palavras desnecessárias, através das chamadas 'stopwords' definidas pela biblioteca.

```
from unicodedata import normalize
wordsList = ' '.join(csv.text.values).split(' ')
wordsList = list(filter(None, wordsList))
wordsFormatted = [normalize('NFKD', word.decode('utf-8')).encode('ASCII', 'ignore').lower() for word in wordsList]
words = pd.DataFrame(wordsFormatted, columns=['word'])
words.head()

stopwords = nltk.corpus.stopwords.words('portuguese')
filteredWords = [word for word in wordsFormatted if word not in stopwords and len(word) >= 3]

words = pd.DataFrame(filteredWords, columns=['word'])
wordsCount = words.word.value_counts().reset_index()
wordsCount.columns = ['word', 'freq']
wordsCount['r'] = wordsCount.freq.rank(ascending=False)
wordsCount.head(10)
```
